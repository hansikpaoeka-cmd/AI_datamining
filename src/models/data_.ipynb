```python
import pandas as pd
train = pd.read_csv(r'I:\PYTHON\AI_datamining\data\raw\train.csv')
train.head(20)
```

```python
train.info()
```

```python
import toad
toad.detect(train)
```

```python
#去除无用特征
test = pd.read_csv(r'I:\PYTHON\AI_datamining\data\raw\test2.csv')
x_train = train.iloc[:,1:]
y_train = train.iloc[:,0]
x_test = test.iloc[:,:-1]
y_test = test.iloc[:,-1]
```

```python
woe = toad.transform.WOETransformer()
x_train_woe = woe.fit_transform(x_train,y_train)

x_test_woe = woe.transform(x_test)
```

```python
y_train.info()
```

```python
toad.detect(x_train_woe)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score,roc_curve


lr_model =LogisticRegression(C=0.1,class_weight='balanced')  
lr_model.fit(x_train_woe,y_train)

y_pred = lr_model.predict_proba(x_train_woe)[:,1]
auc = roc_auc_score(y_train,y_pred)
fpr,tpr,_=roc_curve(y_train,y_pred)
ks = abs(fpr-tpr).max()
print('auc:',auc)
print('ks:',ks)


```

```python
y_pred = lr_model.predict_proba(x_test_woe)[:,1]
auc_t = roc_auc_score(y_test,y_pred)
fpr_t,tpr_t,_=roc_curve(y_test,y_pred)
ks_t = abs(fpr_t-tpr_t).max()
print('auc_t:',auc_t)
print('ks_t:',ks_t)
```
